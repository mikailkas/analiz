```python
# =============================================================================
# GELÄ°ÅMÄ°Å FÄ°NANSAL ANALÄ°Z VE DEÄERLEME SÄ°STEMÄ° (BACKTEST EKLENDÄ°)
# AmaÃ§: Hisse senedi yatÄ±rÄ±m kararlarÄ± iÃ§in kapsamlÄ± analiz + geÃ§miÅŸe dÃ¶nÃ¼k test
# Ã–zellikler: Finansal saÄŸlÄ±k, gelecek performans, deÄŸerleme, hedef fiyat + backtest
# =============================================================================
# 1. KURULUM VE KÃœTÃœPHANELER
# =============================================================================
!pip install aiohttp nest_asyncio yfinance openpyxl pandas numpy matplotlib seaborn scikit-learn -q
import pandas as pd
import numpy as np
import asyncio
import aiohttp
import time
import logging
from datetime import datetime, timedelta
import nest_asyncio
import io
import json
from IPython.display import display, HTML, clear_output
import warnings
from pathlib import Path
import openpyxl
from openpyxl.styles import Font, PatternFill, Alignment, Border, Side, NamedStyle
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.formatting.rule import ColorScaleRule, DataBarRule
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pickle
import os
from scipy import stats
# Ayarlar
warnings.filterwarnings('ignore')
nest_asyncio.apply()
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)
# Stil ayarlarÄ±
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
print("âœ… GeliÅŸmiÅŸ Finansal Analiz Sistemi kuruldu!")
# =============================================================================
# 2. GELÄ°ÅMÄ°Å FÄ°NANSAL ANALÄ°Z SINIFI (BACKTEST Ä°LE GENÄ°ÅLETÄ°LDÄ°)
# =============================================================================
class AdvancedFinancialAnalyzer:
    def __init__(self):
        # Finansal tablo kalemleri - geniÅŸletilmiÅŸ
        self.financial_items = {
            'aktif_toplam': ['Aktif ToplamÄ±', 'AKTIF TOPLAMI', 'Toplam Aktifler', 'TOPLAM VARLIKLAR'],
            'net_satis': ['Net SatÄ±ÅŸlar', 'NET SATIÅLAR', 'SatÄ±ÅŸ Gelirleri', 'HasÄ±lat', 'HASILAT'],
            'satis_maliyeti': ['SatÄ±ÅŸlarÄ±n Maliyeti', 'SATIÅLARIN MALÄ°YETÄ°', 'SatÄ±ÅŸ Maliyeti'],
            'brut_kar': ['BrÃ¼t KÃ¢r', 'BRÃœT KAR', 'BrÃ¼t SatÄ±ÅŸ KarÄ±', 'BRÃœT SATIÅ KARI'],
            'faaliyet_giderleri': ['Faaliyet Giderleri', 'FAALÄ°YET GÄ°DERLERÄ°', 'Genel YÃ¶netim Giderleri'],
            'faaliyet_kari': ['Faaliyet KarÄ±', 'FAALÄ°YET KARI', 'Esas Faaliyet KarÄ±', 'ESAS FAALÄ°YET KARI',
                             'FAALÄ°YET KÃ‚RI', 'ESAS FAALÄ°YET KÃ‚RI'],
            'finansman_gideri': ['Finansman Giderleri', 'FÄ°NANSMAN GÄ°DERLERÄ°'],
            'vergi_oncesi_kar': ['Vergi Ã–ncesi Kar', 'VERGÄ° Ã–NCESÄ° KAR', 'Vergi Ã–ncesi Faaliyet KarÄ±'],
            'net_kar': ['Net Kar', 'NET KAR', 'DÃ¶nem KarÄ±', 'DÃ–NEM KARI', 'Net DÃ¶nem KarÄ±'],
            'ozsermaye': ['Ã–zkaynaklar', 'Ã–ZKAYNAKLAR', 'Toplam Ã–zkaynaklar', 'TOPLAM Ã–ZKAYNAKLAR',
                          'Ã–zkaynaklar ToplamÄ±', 'Ã–ZKAYNAKLAR TOPLAMI', 'Toplam Ã–zkaynak', 'TOPLAM Ã–ZKAYNAK',
                          'Ana OrtaklÄ±ÄŸa Ait Ã–zkaynaklar', 'ANA ORTAKLIÄA AÄ°T Ã–ZKAYNAKLAR'],
            'odenmis_sermaye': ['Ã–denmiÅŸ Sermaye', 'Ã–DENMÄ°Å SERMAYE', 'Ã–denmiÅŸ Sermaye TutarÄ±'],
            'kisa_vadeli_yuk': ['KÄ±sa Vadeli YÃ¼kÃ¼mlÃ¼lÃ¼kler', 'KISA VADELÄ° YÃœKÃœMLÃœLÃœKLER'],
            'uzun_vadeli_yuk': ['Uzun Vadeli YÃ¼kÃ¼mlÃ¼lÃ¼kler', 'UZUN VADELÄ° YÃœKÃœMLÃœLÃœKLER'],
            'toplam_yuk': ['Toplam YÃ¼kÃ¼mlÃ¼lÃ¼kler', 'TOPLAM YÃœKÃœMLÃœLÃœKLER', 'YÃ¼kÃ¼mlÃ¼lÃ¼kler ToplamÄ±'],
            'nakit': ['Nakit ve Nakit Benzerleri', 'NAKÄ°T VE NAKÄ°T BENZERLERÄ°', 'Nakit VarlÄ±klar'],
            'stoklar': ['Stoklar', 'STOKLAR'],
            'ticari_alacaklar': ['Ticari Alacaklar', 'TÄ°CARÄ° ALACAKLAR', 'KÄ±sa Vadeli Ticari Alacaklar'],
            'genel_yonetim_giderleri': ['Genel YÃ¶netim Giderleri (-)','GENEL YÃ–NETÄ°M GÄ°DERLERÄ°'],
            'Pazarlama_SatÄ±ÅŸ_ve_DaÄŸÄ±tÄ±m_Giderleri' :['Pazarlama, SatÄ±ÅŸ ve DaÄŸÄ±tÄ±m Giderleri (-)','PAZARLAMA SATIÅ VE DAÄITIM GÄ°DERLERÄ°'],
            'AraÅŸtÄ±rma_ve_GeliÅŸtirme_Giderleri (-)' : ['AraÅŸtÄ±rma ve GeliÅŸtirme Giderleri (-)','ARAÅTIRMA VE GELÄ°ÅTÄ°RME GÄ°DERLERÄ°'],
            'Amortisman_&_Ä°tfa_PaylarÄ±' : [ 'Amortisman & Ä°tfa PaylarÄ±']
        }
        # SektÃ¶rel gruplar
        self.sektoral_groups = {
            'BANKALAR': ['AKBNK', 'GARAN', 'ISCTR', 'HALKB', 'VAKBN', 'YKBNK'],
            'HOLDÄ°NG': ['KCHOL', 'SAHOL', 'DOHOL', 'AGHOL', 'ALARK', 'ATSYH', 'AVHOL', 'BERA', 'BLCYT', 'BRYAT', 'COSMO', 'DAGHL', 'DENGE', 'YESIL', 'ECZYT', 'ECILC', 'EUHOL', 'GLRYH', 'GLYHO', 'GSDHO', 'IEYHO', 'IHLAS', 'IHYAY', 'ISBIR', 'KERVN', 'MARKA', 'METRO', 'MMCAS', 'MZHLD', 'NTHOL', 'OSTIM', 'OYLUM', 'POLHO', 'RALYH', 'TKFEN', 'UFUK', 'VERUS', 'OTTO', 'INVEO', 'UNLU', 'HEDEF', 'INVES', 'IZINV', 'KLRHO', 'UMPAS', 'LRSHO', 'BINHO'],
            'ENERJÄ° ÃœRETÄ°M': ['AKENR', 'GWIND', 'AKSEN', 'AKSUE', 'ALARK', 'AYEN', 'ZOREN', 'YAYLA', 'ENJSA', 'AYDEM', 'CANTE', 'NATEN', 'ODAS', 'PRKME', 'NTGAZ', 'BIOEN', 'CONSE', 'MAGEN', 'ESEN', 'PAMEL', 'BASGZ', 'MANAS', 'ARASE', 'HUNER', 'ZEDUR', 'AHGAZ', 'AKFYE', 'IZENR', 'TATEN', 'ENERY', 'CATES', 'MOGAN', 'ENTRA', 'LYDYE', 'BIGEN', 'ENDAE', 'A1YEN'],
            'TEKNOLOJÄ° ÃœRÃœN TÄ°CARETÄ°': ['ARENA', 'AZTEK', 'DESPC', 'DGATE', 'INDES', 'MOBTL', 'PENTA', 'TKNSA', 'INGRM'],
            'TAÅ TOPRAK': ['AFYON', 'AKCNS', 'BOBET', 'BASCM', 'BSOKE', 'BTCIM', 'CIMSA', 'CMBTN', 'CMENT', 'GOLTS', 'KONYA', 'NIBAS', 'NUHCM', 'OYAKC', 'BUCIM', 'MARBL', 'LMKDC'],
            'SPOR': ['BJKAS', 'FENER', 'GSRAY', 'TSPOR'],
            'TOPTAN PARAKENDE': ['ARENA', 'AVTUR', 'BIMAS', 'BIZIM', 'CRFSA', 'DESPC', 'DGATE', 'DOAS', 'ETILR', 'INDES', 'INTEM', 'PENTA', 'MEPET', 'MGROS', 'PSDTC', 'SANKO', 'SOKM', 'TGSAS', 'TKNSA', 'KIMMR', 'MOBTL', 'GMTAS', 'AZTEK', 'INGRM', 'EBEBK', 'DCTTR', 'LYDHO'],
            'VARLIK YÃ–NETÄ°MÄ°': ['GLCVY', 'BRKVY', 'SMRVA'],
            'TURÄ°ZM': ['GRSEL', 'AYCES', 'KSTUR', 'MAALT', 'MARTI', 'MERIT', 'METUR', 'GZNMI', 'PKENT', 'ZEDUR', 'TUREX', 'TEKTU', 'ULAS', 'MEPET'],
            'TARIM HAYVANCILIK': ['YAPRK', 'OZSUB', 'DCTTR', 'CEMZY'],
            'SÄ°GORTA': ['AKGRT', 'ANSGR', 'RAYSG', 'TURSG'],
            'OTOMOTÄ°V YAN SANAYÄ°': ['BALAT', 'BFREN', 'BRISA', 'DITAS', 'DOKTA', 'EGEEN', 'FMIZP', 'GOODY', 'JANTS', 'KATMR', 'PARSN', 'YIGIT', 'BAHKM'],
            'SAVUNMA': ['ASELS', 'OTKAR', 'PAPIL', 'SDTTR', 'ALTINY', 'ONRYT', 'KATMR', 'PATEK'],
            'MOBÄ°LYA DEKAROSYON': ['YONGA', 'YATAS', 'ORMA', 'SUMAS', 'DGNMO', 'GENTS', 'KLSYN', 'EUREN'],
            'OTOMOTÄ°V': ['ASUZU', 'DOAS', 'FROTO', 'KARSN', 'OTKAR', 'TMSN', 'TOASO', 'TTRAK', 'ESCAR', 'PLTUR'],
            'MENKUL KIYMET YATIRIM': ['ATLAS', 'ETYAT', 'EUKYO', 'EUYO', 'GRNYO', 'ISYAT', 'MTRYO', 'OYAYO', 'VKFYO'],
            'METAL EÅYA MAKÄ°NE': ['GEREL', 'FORMT', 'KATMR', 'KLMSN', 'BMSCH', 'MAKTK', 'PRKAB', 'SAFKR', 'ULUSE', 'VESBE', 'VESTL', 'ERCB', 'IMASM', 'HKTM', 'SNICA'],
            'KIMYA PLASTÄ°K': ['ACSEL', 'ISKPL', 'AKSA', 'ALKIM', 'KLKIM', 'AYGAZ', 'BAGFS', 'BRISA', 'DYOBY', 'EGGUB', 'EGPRO', 'EKIZ', 'EPLAS', 'GOODY', 'GUBRF', 'HEKTS', 'IZFAS', 'KORDS', 'MEGAP', 'MRSHL', 'PETKM', 'POLHO', 'SANFM', 'SASA', 'SEKUR', 'SEYKM', 'SISE', 'SODSN', 'TMPOL', 'TRCAS', 'TUPRS', 'MERCN', 'CASA', 'ISSEN', 'RNPOL', 'DNISI', 'KMPUR', 'KOPOL', 'TARKM', 'EFORC', 'BAHKM'],
            'MADENCÄ°LÄ°K': ['IPEKE', 'KOZAA', 'KOZAL', 'SARKY', 'CVKMD', 'VSNMD', 'RUZYE'],
            'KAÄIT ÃœRÃœNLERÄ°': ['ALKA', 'TEZOL', 'DOBUR', 'DURDO', 'VKING', 'HURGZ', 'IHGZT', 'KARTN', 'PRZMA', 'SAMAT', 'KONKA', 'MNDTR', 'LILAK'],
            'Ä°NÅAAT': ['ANELE', 'GESAN', 'BRKSN', 'CMBTN', 'YYAPI', 'YBTAS', 'ENKAI', 'IHLGM', 'BRLSM', 'ORGE', 'SANEL', 'TKFEN', 'TURGG', 'PAMEL', 'GLRMK', 'PNLSN', 'EUREN', 'AKFIS'],
            'Ä°LAÃ‡': ['DEVA', 'ECILC', 'GENIL', 'TRILC', 'LKMNH', 'MPARK', 'RTALB', 'SELEC', 'SEYKM', 'EGEPO', 'ANGEN', 'MEDTR', 'ONCSM', 'TNZTP'],
            'Ä°MALAT': ['MRSHL', 'EMKEL', 'KARSN', 'KARTN', 'BOSSA', 'FROTO', 'KRDMA', 'TCKRC', 'KORDS', 'BRISA', 'BURCE', 'SASA', 'AKSA', 'JANTS', 'FMIZP', 'DOFER', 'ARCLK', 'BURVA', 'MEKAG', 'BRKO', 'BRMEN', 'KLMSN', 'BALAT', 'EPLAS', 'KATMR', 'AYGAZ', 'GENTS', 'BNTAS', 'BVSAN', 'GIPTA', 'GUBRF', 'KLKIM', 'KCAER', 'HATSN', 'EREGL', 'AYES', 'RNPOL', 'ALCAR', 'BRSAN', 'DMSAS', 'MAKIM', 'POLTK', 'ALVES', 'FORMT', 'ADEL', 'GEDZA', 'ALKIM', 'ASELS', 'BAYRK', 'ERBOS', 'HKTM', 'BFREN', 'DERHL', 'GEREL', 'DOGUB', 'ARZUM', 'SAFKR', 'PNLSN', 'KAPLM', 'KBORU', 'TOASO', 'DOKTA', 'HEKTS', 'SNICA', 'ORMA', 'CUSAN', 'OZATD', 'IHEVA', 'PRKAB', 'SUMAS', 'YKSLN', 'SILVR', 'ISDMR', 'IZFAS', 'IZDMC', 'TUCLK', 'CEMAS', 'ULUSE', 'CEMTS'],
            'HABERLEÅME': ['TCELL', 'ALCTL', 'KRONT', 'NETAS', 'TTKOM', 'IHAAS'],
            'GÄ°YÄ°M TEKSTÄ°L': ['DAGI', 'DERIM', 'DESA', 'MAVI', 'MNDRS', 'RODRG', 'ROYAL', 'SANKO', 'SUWEN', 'VAKKO', 'KOTON', 'DERHL', 'ARSAN', 'ATEKS', 'BAYRK', 'BOSSA', 'BRKO', 'BRMEN', 'DIRIT', 'YUNSA', 'HATEK', 'KRTEK', 'LUKSK', 'RODRG', 'SKTAS', 'SNPAM', 'ENSRI', 'SUNTK', 'RUBNS', 'BLCYT', 'SONME', 'ARTMS', 'SANKO'],
            'GÄ°RÄ°ÅÄ°M SERMAYESÄ°': ['GOZDE', 'HDFGS', 'HUBVC', 'ISGSY', 'VERTU', 'PRDGS', 'ICUGS', 'BULGS'],
            'GAYRÄ°MENKUL': ['AGYO', 'AKFGY', 'AKMGY', 'AKSGY', 'ALGYO', 'ATAGY', 'AVGYO', 'DGGYO', 'DZGYO', 'YGYO', 'YGGYO', 'EKGYO', 'HLGYO', 'IDGYO', 'IHLGM', 'ISGYO', 'KLGYO', 'KRGYO', 'KUYAS', 'MRGYO', 'MSGYO', 'NUGYO', 'OZGYO', 'OZKY', 'PAGYO', 'PEKGY', 'RYGYO', 'SNGYO', 'SRVGY', 'TDGYO', 'TRGYO', 'ZRGYO', 'TSGYO', 'VKGYO', 'KZBGY', 'PSGYO', 'KGYO', 'DAPGM', 'SEGYO', 'ADESE', 'EYGYO', 'FZLGY', 'ASGYO', 'KZGYO', 'ADGYO', 'MHRGY', 'VRGYO', 'BEGYO', 'SURGY', 'AVPGY', 'RGYAS', 'PEHOL', 'LYDHO', 'AHSGY', 'EDIP', 'EGEGY'],
            'FÄ°N. KÄ°RALAMA': ['ISFIN', 'QNBFL', 'SEKFK', 'VAKFN', 'QNBFK'],
            'FAKTORÄ°NG': ['CRDFA', 'GARFA', 'LIDFA', 'ULUFA', 'DSTFK'],
            'ENERJÄ° TEKNOLOJÄ°': ['SMRTG', 'ALFAS', 'ASTOR', 'EUPWR', 'CWENE', 'GESAN', 'YEOTK', 'KONTR', 'BRLSM', 'ORGE', 'SAYAS', 'EKOS', 'ONRYT', 'KLYPV'],
            'EMEKLÄ°LÄ°K': ['ANHYT', 'AGESA'],
            'DESTEK VE HÄ°ZMET': ['BEYAZ', 'CEOEM', 'AKYHO', 'FLAP', 'SNKRN', 'MACKO', 'HRKET', 'PCILT', 'DOCO'],
            'DAYANIKLI TÃœKETÄ°M MALLARI': ['ALCAR', 'ARCLK', 'SILVR', 'VESBE', 'VESTL', 'ARZUM', 'IHEVA', 'KLMSN'],
            'CAM SERAMÄ°K': ['QUAGR', 'EGSER', 'KUTPO', 'SISE', 'USAK', 'BIENY', 'KLSER', 'CGCAM', 'SERNT'],
            'BÄ°LÄ°ÅÄ°M VE YAZILIM': ['ARDYZ', 'ARENA', 'ATATP', 'VBTYZ', 'DESPC', 'DGATE', 'ESCOM', 'FONET', 'KONTR', 'MTRKS', 'INDES', 'PENTA', 'KAREL', 'KFEIN', 'KRONT', 'LINK', 'LOGO', 'NETAS', 'PAPIL', 'PKART', 'SMART', 'EDATA', 'MIATK', 'MOBTL', 'YEOTK', 'HTTBT', 'OBASE', 'AZTEK', 'INGRM', 'FORTE', 'REEDR', 'PATEK', 'ODINE', 'SKYLP', 'ONRYT', 'INTEK'],
            'ARACI KURUM': ['OYYAT', 'GEDIK', 'GLBMD', 'INFO', 'ISMEN', 'OSMEN', 'TERA', 'A1CAP', 'SKYMD'],
            'ANA METAL': ['BRSAN', 'BURCE', 'BURVA', 'CELHA', 'CEMAS', 'CEMTS', 'CUSAN', 'DMSAS', 'DOKTA', 'YKSLN', 'ERBOS', 'EREGL', 'ISDMR', 'IZMDC', 'KRDMA', 'KRDMB', 'KRDMD', 'BMSCH', 'POLTK', 'SARKY', 'TUCLK', 'ERCB', 'BMSTL', 'KCAER', 'KBORU', 'MEGMT', 'KOCMT', 'OZYSR'],
            'AMBALAJ': ['BNTAS', 'BAKAB', 'DURDO', 'EMNIS', 'GEDZA', 'KAPLM', 'SEKUR', 'KRPLS', 'BARMA', 'MNDTR', 'OZRDN'],
            'GIDA PERAKENDE': ['MGROS', 'VESTL', 'BIMAS', 'BIZIM', 'CRFSA', 'GMTAS', 'SOKM', 'KIMMR', 'MOPAS'],
            'GIDA VE Ä°Ã‡ECEK': ['AEFES', 'AVOD', 'BANVT', 'ORCAY', 'CCOLA', 'DARDL', 'ARMGD', 'ERSU', 'FADE', 'FRIGO', 'KENT', 'KNFRT', 'KRSTL', 'MERKO', 'PENGD', 'PETUN', 'PINSU', 'SELGD', 'TATGD', 'TBORG', 'TUKAS', 'ULKER', 'ULUUN', 'VANGD', 'KRVGD', 'SELVA', 'ELITE', 'YYLGD', 'KUVVA', 'SOKE', 'GOKNR', 'EKSUN', 'BIGCH', 'KAYSE', 'ATAKP', 'OFSYM', 'BYDNR', 'OYLUM', 'DMRGD', 'TABGD', 'AGROT', 'BORSK', 'OBAMS', 'ALKLC', 'SEGMN', 'EFORC', 'GUNDG', 'CEMZY', 'DURKN', 'GRTHO', 'BALSU', 'BESLR'],
            'ULAÅTIRMA': ['THYAO', 'TAVHL', 'CLEBI', 'GSDDE', 'LIDER', 'PGSUS', 'RYSAS', 'TLMAN', 'PLTUR', 'PASEU', 'HOROZ', 'HRKET', 'BINBN']
        }
        # Hesaplama formÃ¼lleri - geniÅŸletilmiÅŸ
        self.ratios = {
            'roe': lambda data: self.safe_divide(data.get('net_kar', 0), data.get('ozsermaye', 0)) * 100,
            'roa': lambda data: self.safe_divide(data.get('net_kar', 0), data.get('aktif_toplam', 0)) * 100,
            'ROIC': lambda d: self.safe_divide(d.get('faaliyet_kari', 0),d.get('aktif_toplam', 0) - d.get('kisa_vadeli_yuk', 0)) * 100,
            'EBITDA_MARJI': lambda d: self.safe_divide(
                d.get('faaliyet_kari', 0) + d.get('Amortisman_&_Ä°tfa_PaylarÄ±', 0),
                d.get('net_satis', 0)
            ) * 100,
            'net_kar_marji': lambda data: self.safe_divide(data.get('net_kar', 0), data.get('net_satis', 0)) * 100,
            'brut_kar_marji': lambda data: self.safe_divide(data.get('brut_kar', 0), data.get('net_satis', 0)) * 100,
            'faaliyet_kar_marji': lambda data: self.safe_divide(data.get('faaliyet_kari', 0), data.get('net_satis', 0)) * 100,
            'aktif_devir_hizi': lambda data: self.safe_divide(data.get('net_satis', 0), data.get('aktif_toplam', 0)),
            'ozsermaye_carpani': lambda data: self.safe_divide(data.get('aktif_toplam', 0), data.get('ozsermaye', 0)),
            'borc_orani': lambda data: self.safe_divide((data.get('kisa_vadeli_yuk', 0) + data.get('uzun_vadeli_yuk', 0)), data.get('aktif_toplam', 0)) * 100,
            'likidite_orani': lambda data: self.safe_divide(
                data.get('nakit', 0) + data.get('ticari_alacaklar', 0) + data.get('stoklar', 0),
                data.get('kisa_vadeli_yuk', 0)
            ),
            'giderler': lambda data: data.get('genel_yonetim_giderleri', 0) +
                                    data.get('Pazarlama_SatÄ±ÅŸ_ve_DaÄŸÄ±tÄ±m_Giderleri', 0) +
                                    data.get('AraÅŸtÄ±rma_ve_GeliÅŸtirme_Giderleri (-)', 0),
            'nakit_orani': lambda data: self.safe_divide(data.get('nakit', 0), data.get('kisa_vadeli_yuk', 0)),
            'asit_test': lambda data: self.safe_divide(
                data.get('nakit', 0) + data.get('ticari_alacaklar', 0),
                data.get('kisa_vadeli_yuk', 0)
            ),
            'pd_dd': lambda data: self.safe_divide(data.get('market_cap', 0), data.get('ozsermaye', 0)),
            'fk': lambda data: self.safe_divide(data.get('current_price', 0), data.get('eps', 0)),
            'ev_ebitda': lambda data: self.safe_divide(data.get('market_cap', 0), data.get('faaliyet_kari', 0) + data.get('Amortisman_&_Ä°tfa_PaylarÄ±', 0)),
            'sermaye_karliligi': lambda data: self.safe_divide(data.get('net_kar', 0), data.get('odenmis_sermaye', 0)) * 100,
            'faaliyet_kari_marji': lambda data: self.safe_divide(data.get('faaliyet_kari', 0), data.get('net_satis', 0)) * 100,
        }
        # Ã–nbellek ve rate limiting iÃ§in deÄŸiÅŸkenler
        self.cache = {}
        self.cache_file = "financial_cache.pkl"
        self.last_call_time = {}
        self.rate_limit_delay = 0.5  # 500ms between API calls
        self.results = []

        # Ã–nbelleÄŸi yÃ¼kle
        self.load_cache()

    def load_cache(self):
        """Ã–nbelleÄŸi dosyadan yÃ¼kle"""
        try:
            if os.path.exists(self.cache_file):
                with open(self.cache_file, 'rb') as f:
                    self.cache = pickle.load(f)
                logger.info(f"Ã–nbellek yÃ¼klendi: {len(self.cache)} kayÄ±t")
        except Exception as e:
            logger.error(f"Ã–nbellek yÃ¼klenemedi: {str(e)}")
            self.cache = {}
    def save_cache(self):
        """Ã–nbelleÄŸi dosyaya kaydet"""
        try:
            with open(self.cache_file, 'wb') as f:
                pickle.dump(self.cache, f)
        except Exception as e:
            logger.error(f"Ã–nbellek kaydedilemedi: {str(e)}")

    def safe_divide(self, numerator, denominator):
        """GÃ¼venli bÃ¶lme iÅŸlemi"""
        if denominator == 0 or pd.isna(denominator) or pd.isna(numerator):
            return 0
        try:
            return float(numerator) / float(denominator)
        except:
            return 0
    def validate_financial_data(self, data):
        """Finansal verilerin tutarlÄ±lÄ±ÄŸÄ±nÄ± kontrol et"""
        warnings = []
        # Muhasebe dengesi kontrolÃ¼
        if data.get('aktif_toplam', 0) < (data.get('ozsermaye', 0) + data.get('toplam_yuk', 0)):
            warnings.append("âš ï¸ Muhasebe dengesi uyuÅŸmuyor: Aktif â‰  Ã–zkaynak + YÃ¼kÃ¼mlÃ¼lÃ¼kler")
        # SÄ±fÄ±r satÄ±ÅŸla kar kontrolÃ¼
        if data.get('net_satis', 0) == 0 and data.get('net_kar', 0) > 0:
            warnings.append("âš ï¸ SÄ±fÄ±r satÄ±ÅŸla kar elde edilmiÅŸ")
        # Negatif Ã¶zkaynak kontrolÃ¼
        if data.get('ozsermaye', 0) < 0:
            warnings.append("âš ï¸ Negatif Ã¶zkaynak tespit edildi")
        # YÃ¼ksek borÃ§ oranÄ± kontrolÃ¼
        borc_orani = self.safe_divide(
            data.get('kisa_vadeli_yuk', 0) + data.get('uzun_vadeli_yuk', 0),
            data.get('aktif_toplam', 0)
        ) * 100
        if borc_orani > 80:
            warnings.append(f"âš ï¸ Ã‡ok yÃ¼ksek borÃ§ oranÄ±: %{borc_orani:.1f}")
        return warnings
    async def get_financial_data_api(self, session, ticker, period_config):
        """API'den finansal veri Ã§eker - rate limiting ile"""
        # Rate limiting kontrolÃ¼
        now = time.time()
        if ticker in self.last_call_time and (now - self.last_call_time[ticker]) < self.rate_limit_delay:
            await asyncio.sleep(self.rate_limit_delay - (now - self.last_call_time[ticker]))
        # Ã–nbellek kontrolÃ¼
        cache_key = f"{ticker}_{period_config['year1']}_{period_config['period1']}"
        if cache_key in self.cache:
            logger.debug(f"{ticker} iÃ§in Ã¶nbellekten veri alÄ±ndÄ±")
            return self.cache[cache_key]
        url = (f"https://www.isyatirim.com.tr/_layouts/15/IsYatirim.Website/Common/Data.aspx/MaliTablo"
               f"?companyCode={ticker}&exchange=TRY&financialGroup=XI_29"
               f"&year1={period_config['year1']}&period1={period_config['period1']}"
               f"&year2={period_config['year2']}&period2={period_config['period2']}"
               f"&year3={period_config['year3']}&period3={period_config['period3']}"
               f"&year4={period_config['year4']}&period4={period_config['period4']}"
               f"&_={int(time.time() * 1000)}")
        try:
            async with session.get(url, timeout=30) as response:
                self.last_call_time[ticker] = time.time()
                if response.status == 200:
                    json_data = await response.json()
                    # Ã–nbelleÄŸe kaydet
                    self.cache[cache_key] = json_data
                    return json_data
                else:
                    logger.error(f"{ticker} iÃ§in API hatasÄ±: {response.status}")
                    return None
        except aiohttp.ClientError as e:
            logger.error(f"{ticker} API baÄŸlantÄ± hatasÄ±: {str(e)}")
            return self.get_cached_data(ticker, period_config)
        except Exception as e:
            logger.error(f"{ticker} API hatasÄ±: {str(e)}")
            return None
    def get_cached_data(self, ticker, period_config):
        """Ã–nbellekten veri dÃ¶ndÃ¼r"""
        cache_key = f"{ticker}_{period_config['year1']}_{period_config['period1']}"
        return self.cache.get(cache_key, None)
    async def get_market_data(self, ticker, historical_date=None):
        """Yahoo Finance Ã¼zerinden piyasa deÄŸeri ve EPS verilerini Ã§eker"""
        try:
            stock = yf.Ticker(f"{ticker}.IS")
            if historical_date:
                # Tarih aralÄ±ÄŸÄ±nÄ± belirle
                end_date = datetime.strptime(historical_date, '%Y-%m-%d')
                start_date = end_date - timedelta(days=365)
                # GeÃ§miÅŸ verileri al
                hist = stock.history(start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'))
                if hist.empty:
                    logger.warning(f"{ticker} iÃ§in {historical_date} tarihinde veri bulunamadÄ±")
                    return {
                        'market_cap': 0,
                        'eps': 0,
                        'current_price': 0,
                        'book_value_per_share': 0,
                        'price_trend': 0
                    }
                # En son tarihteki verileri al
                last_row = hist.iloc[-1]
                current_price = last_row['Close']
                # Piyasa deÄŸeri tahmini (o dÃ¶nemdeki hisse sayÄ±sÄ± * fiyat)
                # Not: Bu bir tahmindir, kesin bilgi iÃ§in daha detaylÄ± veri gerekir
                shares_outstanding = stock.info.get('sharesOutstanding', 0)
                if shares_outstanding == 0:
                    # Alternatif yÃ¶ntem: mevcut piyasa deÄŸeri ve mevcut fiyatÄ± kullan
                    current_market_cap = stock.info.get('marketCap', 0)
                    current_price_now = stock.info.get('currentPrice', 0)
                    if current_price_now > 0:
                        shares_outstanding = current_market_cap / current_price_now
                market_cap = shares_outstanding * current_price / 1000000  # Milyon TL'ye Ã§evir
                # EPS ve defter deÄŸeri iÃ§in mevcut verileri kullan (geÃ§miÅŸ veriler sÄ±nÄ±rlÄ±)
                eps = stock.info.get('trailingEps', 0)
                book_value_per_share = stock.info.get('bookValue', 0)
                # Fiyat trendini hesapla
                price_trend = self.calculate_price_trend(hist)
                return {
                    'market_cap': market_cap,
                    'eps': eps,
                    'current_price': current_price,
                    'book_value_per_share': book_value_per_share,
                    'price_trend': price_trend
                }
            else:
                # Mevcut verileri al
                info = stock.info
                market_cap = info.get('marketCap', 0) / 1000000  # Milyon TL'ye Ã§evir
                eps = info.get('trailingEps', 0)
                current_price = info.get('currentPrice', 0)
                book_value_per_share = info.get('bookValue', 0)
                # GeÃ§miÅŸ fiyat verileri (trend analizi iÃ§in)
                hist = stock.history(period="1y")
                price_trend = self.calculate_price_trend(hist)
                return {
                    'market_cap': market_cap,
                    'eps': eps,
                    'current_price': current_price,
                    'book_value_per_share': book_value_per_share,
                    'price_trend': price_trend
                }
        except Exception as e:
            logger.error(f"{ticker} iÃ§in piyasa verisi alÄ±namadÄ±: {str(e)}")
            return {
                'market_cap': 0,
                'eps': 0,
                'current_price': 0,
                'book_value_per_share': 0,
                'price_trend': 0
            }
    def calculate_price_trend(self, hist_data):
        """Fiyat trendini hesapla (yÄ±llÄ±k deÄŸiÅŸim %)"""
        if len(hist_data) < 2:
            return 0
        try:
            first_price = hist_data['Close'].iloc[0]
            last_price = hist_data['Close'].iloc[-1]
            return ((last_price - first_price) / first_price) * 100
        except:
            return 0
    def extract_financial_values(self, df, period_col):
        """DataFrame'den finansal deÄŸerleri Ã§Ä±karÄ±r"""
        values = {}
        found_items = []
        for key, possible_names in self.financial_items.items():
            value = 0
            found = False
            for name in possible_names:
                matching_rows = df[df['Kalem'].str.contains(name, case=False, na=False, regex=False)]
                if not matching_rows.empty:
                    val = matching_rows.iloc[0][period_col]
                    if pd.notna(val) and val != 0:
                        value = float(val)
                        found = True
                        found_items.append(f"{key}: {name} = {value}")
                        break
            if not found and key == 'faaliyet_kari':
                logger.warning(f"Faaliyet kÃ¢rÄ± bulunamadÄ±. Kalemler: {df['Kalem'].tolist()}")
            values[key] = value
        logger.debug(f"Bulunan finansal kalemler: {found_items}")
        return values
    def calculate_ratios(self, financial_data):
        """Finansal oranlarÄ± hesaplar"""
        ratios = {}
        for ratio_name, formula in self.ratios.items():
            try:
                ratios[ratio_name] = formula(financial_data)
            except Exception as e:
                ratios[ratio_name] = 0
                logger.error(f"{ratio_name} oranÄ± hesaplanamadÄ±: {str(e)}")
        return ratios
    def calculate_financial_health_score(self, data):
        """Finansal saÄŸlÄ±k skorunu hesapla (0-100)"""
        score = 0
        details = {}
        # ROE Skoru (0-25)
        roe = data.get('roe', 0)
        if roe > 20:
            score += 25
            details['ROE'] = f"25/25 (Ã‡ok iyi: %{roe:.1f})"
        elif roe > 15:
            score += 20
            details['ROE'] = f"20/25 (Ä°yi: %{roe:.1f})"
        elif roe > 10:
            score += 15
            details['ROE'] = f"15/25 (Orta: %{roe:.1f})"
        elif roe > 5:
            score += 10
            details['ROE'] = f"10/25 (ZayÄ±f: %{roe:.1f})"
        else:
            details['ROE'] = f"0/25 (Ã‡ok zayÄ±f: %{roe:.1f})"
        # BorÃ§ OranÄ± Skoru (0-20)
        borc_orani = data.get('borc_orani', 100)
        if borc_orani < 30:
            score += 20
            details['BorÃ§'] = f"20/20 (Ã‡ok dÃ¼ÅŸÃ¼k: %{borc_orani:.1f})"
        elif borc_orani < 50:
            score += 15
            details['BorÃ§'] = f"15/20 (DÃ¼ÅŸÃ¼k: %{borc_orani:.1f})"
        elif borc_orani < 70:
            score += 10
            details['BorÃ§'] = f"10/20 (Orta: %{borc_orani:.1f})"
        else:
            details['BorÃ§'] = f"0/20 (YÃ¼ksek: %{borc_orani:.1f})"
        # Likidite Skoru (0-20)
        likidite = data.get('likidite_orani', 0)
        if likidite > 2:
            score += 20
            details['Likidite'] = f"20/20 (Ã‡ok iyi: {likidite:.1f})"
        elif likidite > 1.5:
            score += 15
            details['Likidite'] = f"15/20 (Ä°yi: {likidite:.1f})"
        elif likidite > 1:
            score += 10
            details['Likidite'] = f"10/20 (Orta: {likidite:.1f})"
        else:
            details['Likidite'] = f"0/20 (DÃ¼ÅŸÃ¼k: {likidite:.1f})"
        # Kar MarjÄ± Skoru (0-20)
        kar_marji = data.get('net_kar_marji', 0)
        if kar_marji > 15:
            score += 20
            details['Kar MarjÄ±'] = f"20/20 (Ã‡ok iyi: %{kar_marji:.1f})"
        elif kar_marji > 10:
            score += 15
            details['Kar MarjÄ±'] = f"15/20 (Ä°yi: %{kar_marji:.1f})"
        elif kar_marji > 5:
            score += 10
            details['Kar MarjÄ±'] = f"10/20 (Orta: %{kar_marji:.1f})"
        else:
            details['Kar MarjÄ±'] = f"0/20 (DÃ¼ÅŸÃ¼k: %{kar_marji:.1f})"
        # BÃ¼yÃ¼me Skoru (0-15)
        price_trend = data.get('price_trend', 0)
        if price_trend > 50:
            score += 15
            details['BÃ¼yÃ¼me'] = f"15/15 (Ã‡ok hÄ±zlÄ±: %{price_trend:.1f})"
        elif price_trend > 20:
            score += 12
            details['BÃ¼yÃ¼me'] = f"12/15 (HÄ±zlÄ±: %{price_trend:.1f})"
        elif price_trend > 0:
            score += 8
            details['BÃ¼yÃ¼me'] = f"8/15 (Pozitif: %{price_trend:.1f})"
        else:
            details['BÃ¼yÃ¼me'] = f"0/15 (Negatif: %{price_trend:.1f})"
        return {
            'score': min(score, 100),
            'details': details,
            'category': self.get_score_category(score)
        }
    def get_score_category(self, score):
        """Skora gÃ¶re kategori belirle"""
        if score >= 80:
            return "Ã‡OK Ä°YÄ°"
        elif score >= 60:
            return "Ä°YÄ°"
        elif score >= 40:
            return "ORTA"
        elif score >= 20:
            return "ZAYIF"
        else:
            return "Ã‡OK ZAYIF"
    def calculate_valuation_score(self, data):
        """DeÄŸerleme skorunu hesapla (0-100) - Ne kadar ucuz? - DÃœZELTÄ°LMÄ°Å VERSÄ°YON"""
        score = 100  # BaÅŸlangÄ±Ã§ta tam puan, indirimler uygulanacak
        details = {}

        # F/K OranÄ± Ä°ndirimi - DÃœZELTÄ°LDÄ°: DÃ¼ÅŸÃ¼k F/K = Ucuz = Pozitif
        fk = data.get('fk', 0)
        if fk <= 0:  # Negatif veya sÄ±fÄ±r F/K
            score -= 50
            details['F/K'] = f"-50 (GeÃ§ersiz F/K: {fk:.1f})"
        elif fk < 5:  # Ã‡ok ucuz
            score += 15
            details['F/K'] = f"+15 (Ã‡ok ucuz: {fk:.1f})"
            score = min(score, 100)
        elif fk < 10:  # Ucuz
            score += 5
            details['F/K'] = f"+5 (Ucuz: {fk:.1f})"
            score = min(score, 100)
        elif fk < 15:  # Hafif pahalÄ±
            score -= 5
            details['F/K'] = f"-5 (Hafif pahalÄ±: {fk:.1f})"
        elif fk < 20:  # PahalÄ±
            score -= 15
            details['F/K'] = f"-15 (PahalÄ±: {fk:.1f})"
        elif fk < 30:  # Ã‡ok pahalÄ±
            score -= 25
            details['F/K'] = f"-25 (Ã‡ok pahalÄ±: {fk:.1f})"
        else:  # AÅŸÄ±rÄ± pahalÄ±
            score -= 40
            details['F/K'] = f"-40 (AÅŸÄ±rÄ± pahalÄ±: {fk:.1f})"

        # PD/DD Ä°ndirimi - DÃœZELTÄ°LDÄ°: DÃ¼ÅŸÃ¼k PD/DD = Ucuz = Pozitif
        pd_dd = data.get('pd_dd', 0)
        if pd_dd <= 0:  # Negatif veya sÄ±fÄ±r PD/DD
            score -= 40
            details['PD/DD'] = f"-40 (GeÃ§ersiz PD/DD: {pd_dd:.1f})"
        elif pd_dd < 2:  # Ã‡ok ucuz
            score += 25
            details['PD/DD'] = f"+25 (Ã‡ok ucuz: {pd_dd:.1f})"
            score = min(score, 100)
        elif pd_dd < 3:  # Ucuz
            score += 10
            details['PD/DD'] = f"+10 (Ucuz: {pd_dd:.1f})"
            score = min(score, 100)
        elif pd_dd < 4:  # Hafif pahalÄ±
            score += 5
            details['PD/DD'] = f"+5 (Hafif pahalÄ±: {pd_dd:.1f})"
        elif pd_dd < 7:  # Orta pahalÄ±
            score -= 5
            details['PD/DD'] = f"-5 (Orta pahalÄ±: {pd_dd:.1f})"
        elif pd_dd < 10:  # PahalÄ±
            score -= 10
            details['PD/DD'] = f"-10 (PahalÄ±: {pd_dd:.1f})"
        else:  # Ã‡ok pahalÄ±
            score -= 25
            details['PD/DD'] = f"-30 (Ã‡ok pahalÄ±: {pd_dd:.1f})"

        # EV/EBITDA Ä°ndirimi - DÃœZELTÄ°LDÄ°: DÃ¼ÅŸÃ¼k EV/EBITDA = Ucuz = Pozitif
        ev_ebitda = data.get('ev_ebitda', 0)
        if ev_ebitda <= 0:  # Negatif veya sÄ±fÄ±r EV/EBITDA
            score -= 30
            details['EV/EBITDA'] = f"-30 (GeÃ§ersiz EV/EBITDA: {ev_ebitda:.1f})"
        elif ev_ebitda < 3:  # Ã‡ok ucuz
            score += 25
            details['EV/EBITDA'] = f"+10 (Ã‡ok ucuz: {ev_ebitda:.1f})"
            score = min(score, 100)
        elif ev_ebitda < 6:  # Ucuz
            score += 15
            details['EV/EBITDA'] = f"+5 (Ucuz: {ev_ebitda:.1f})"
            score = min(score, 100)
        elif ev_ebitda < 10:  # Hafif pahalÄ±
            score += 10
            details['EV/EBITDA'] = f"+10 (Hafif pahalÄ±: {ev_ebitda:.1f})"
        elif ev_ebitda < 15:  # PahalÄ±
            score -= 10
            details['EV/EBITDA'] = f"-8 (PahalÄ±: {ev_ebitda:.1f})"
        elif ev_ebitda < 20:  # Ã‡ok pahalÄ±
            score -= 15
            details['EV/EBITDA'] = f"-15 (Ã‡ok pahalÄ±: {ev_ebitda:.1f})"
        else:  # AÅŸÄ±rÄ± pahalÄ±
            score -= 25
            details['EV/EBITDA'] = f"-25 (AÅŸÄ±rÄ± pahalÄ±: {ev_ebitda:.1f})"

        return {
            'score': max(score, 0),
            'details': details,
            'category': self.get_score_category(score)
        }
    def calculate_future_performance_score(self, data):
        """Gelecek performans skorunu hesapla (0-100)"""
        score = 0
        details = {}
        # ROE Trendi
        roe = data.get('roe', 0)
        if roe > 50:
            score += 30
            details['ROE'] = f"30/30 (SÃ¼rdÃ¼rÃ¼lebilir yÃ¼ksek kÃ¢rlÄ±lÄ±k: %{roe:.1f})"
        elif roe > 25:
            score += 20
            details['ROE'] = f"25/30 (Ä°yi kÃ¢rlÄ±lÄ±k: %{roe:.1f})"
        elif roe > 15:
            score += 10
            details['ROE'] = f"15/30 (Orta kÃ¢rlÄ±lÄ±k: %{roe:.1f})"
        elif roe > 5:
            score += 5
            details['ROE'] = f"5/30 (Orta kÃ¢rlÄ±lÄ±k: %{roe:.1f})"
        else:
            details['ROE'] = f"0/30 (DÃ¼ÅŸÃ¼k kÃ¢rlÄ±lÄ±k: %{roe:.1f})"
        # YatÄ±rÄ±m OranÄ±
        yatirim_orani = self.safe_divide(
            data.get('ozsermaye', 0) - data.get('odenmis_sermaye', 0),
            data.get('ozsermaye', 0)
        ) * 100
        if yatirim_orani > 50:
            score += 25
            details['YatÄ±rÄ±m'] = f"25/25 (YÃ¼ksek sermaye birikimi: %{yatirim_orani:.1f})"
        elif yatirim_orani > 30:
            score += 20
            details['YatÄ±rÄ±m'] = f"20/25 (Orta sermaye birikimi: %{yatirim_orani:.1f})"
        elif yatirim_orani > 10:
            score += 10
            details['YatÄ±rÄ±m'] = f"10/25 (DÃ¼ÅŸÃ¼k sermaye birikimi: %{yatirim_orani:.1f})"
        else:
            details['YatÄ±rÄ±m'] = f"0/25 (SÄ±nÄ±rlÄ± sermaye birikimi: %{yatirim_orani:.1f})"
        # Faaliyet KÃ¢rÄ± Trendi
        faaliyet_kari = data.get('faaliyet_kari', 0)
        if faaliyet_kari > 0:
            score += 20
            details['Faaliyet KÃ¢rÄ±'] = f"20/20 (Pozitif faaliyet kÃ¢rÄ±: {faaliyet_kari:.1f}M TL)"
        else:
            details['Faaliyet KÃ¢rÄ±'] = f"0/20 (Negatif faaliyet kÃ¢rÄ±: {faaliyet_kari:.1f}M TL)"
        # SektÃ¶r BÃ¼yÃ¼me Potansiyeli (basit bir yaklaÅŸÄ±m)
        sektor = data.get('sektor', 'BILINMIYOR')
        yuksek_potansiyel_sektorler = ['TEKNOLOJÄ°', 'ENERJÄ°', 'SAVUNMA', 'BÄ°LÄ°ÅÄ°M', 'Ä°LAÃ‡']
        orta_potansiyel_sektorler = ['HOLDÄ°NG', 'OTOMOTÄ°V', 'GIDA', 'PERAKENDE']
        if any(sektor.startswith(s) for s in yuksek_potansiyel_sektorler):
            score += 25
            details['SektÃ¶r'] = f"25/25 (YÃ¼ksek bÃ¼yÃ¼me potansiyeli: {sektor})"
        elif any(sektor.startswith(s) for s in orta_potansiyel_sektorler):
            score += 15
            details['SektÃ¶r'] = f"15/25 (Orta bÃ¼yÃ¼me potansiyeli: {sektor})"
        else:
            score += 5
            details['SektÃ¶r'] = f"5/25 (DÃ¼ÅŸÃ¼k bÃ¼yÃ¼me potansiyeli: {sektor})"
        return {
            'score': min(score, 100),
            'details': details,
            'category': self.get_score_category(score)
        }
    def calculate_sector_averages(self, results):
        """SektÃ¶r ortalamalarÄ±nÄ± hesapla"""
        sector_pe = {}
        sector_pb = {}
        for sector in set(r['SEKTÃ–R'] for r in results):
            pe_values = []
            pb_values = []
            for r in results:
                if r['SEKTÃ–R'] == sector:
                    if r['F/K'] > 0:
                        pe_values.append(r['F/K'])
                    if r['PD/DD'] > 0:
                        pb_values.append(r['PD/DD'])
            sector_pe[sector] = np.mean(pe_values) if pe_values else 15.0
            sector_pb[sector] = np.mean(pb_values) if pb_values else 1.5
        return sector_pe, sector_pb
    def calculate_target_price(self, result, sector_pe_avg, sector_pb_avg):
        """Hedef fiyat hesapla - SEKTÃ–R ORTALAMALARI KULLANILMADAN"""
        # Gerekli verileri al
        eps = result.get('EPS', 0)
        book_value = result.get('DEFTER DEÄERÄ°', 0)
        current_price = result.get('FÄ°YAT', 0)
        future_score = result.get('GELECEK SKORU', 0)

        # GeÃ§ersiz veri kontrolÃ¼
        if eps <= 0 or book_value <= 0 or current_price <= 0:
            return {
                'HEDEF FÄ°YAT': 0,
                'YÃœKSELÄ°Å POTANSÄ°YELÄ° (%)': 0,
                'HEDEF FÄ°YAT AÃ‡IKLAMA': 'Yetersiz veri'
            }

        # BÃ¼yÃ¼me Ã§arpanÄ±
        if future_score >= 80:
            growth_multiplier = 1.2
        elif future_score >= 60:
            growth_multiplier = 1.1
        elif future_score < 40:
            growth_multiplier = 0.9
        else:
            growth_multiplier = 1.0

        # Standart Ã§arpanlar (sektÃ¶r ortalamalarÄ± yerine)
        standard_pe = 12.0  # Standart F/K Ã§arpanÄ±
        standard_pb = 1.8   # Standart PD/DD Ã§arpanÄ±

        # Hedef fiyat hesaplamalarÄ±
        target_pe = eps * standard_pe * growth_multiplier
        target_pb = book_value * standard_pb * growth_multiplier

        # AÄŸÄ±rlÄ±klÄ± ortalama (60% PE, 40% PB)
        target_price = (target_pe * 0.6 + target_pb * 0.4)

        # YÃ¼kseliÅŸ potansiyeli
        upside_potential = ((target_price - current_price) / current_price) * 100

        return {
            'HEDEF FÄ°YAT': round(target_price, 2),
            'YÃœKSELÄ°Å POTANSÄ°YELÄ° (%)': round(upside_potential, 1),
            'HEDEF FÄ°YAT AÃ‡IKLAMA': f"Standart F/K: {standard_pe:.1f}, Standart PD/DD: {standard_pb:.1f}, BÃ¼yÃ¼me Ã‡arpanÄ±: {growth_multiplier:.1f}x"
        }
    async def analyze_ticker(self, session, ticker, semaphore, historical_date=None):
        """Tek bir hisse iÃ§in kapsamlÄ± analiz yapar"""
        async with semaphore:
            try:
                # GÃ¼ncel ve geÃ§miÅŸ dÃ¶nem verileri
                if historical_date:
                    # Tarih bilgilerini ayarla
                    hist_date = datetime.strptime(historical_date, '%Y-%m-%d')
                    year1 = hist_date.year
                    month1 = hist_date.month
                    # DÃ¶nem belirleme (3 aylÄ±k periyotlar)
                    if month1 <= 3:
                        period1 = 3
                    elif month1 <= 6:
                        period1 = 6
                    elif month1 <= 9:
                        period1 = 9
                    else:
                        period1 = 12
                    # Ã–nceki dÃ¶nemler
                    year2 = year1
                    period2 = period1 - 3 if period1 > 3 else 12
                    if period2 == 12:
                        year2 = year1 - 1
                    year3 = year1 - 1
                    period3 = 12
                    year4 = year1 - 1
                    period4 = 9
                    period_config = {
                        "year1": year1, "period1": period1,
                        "year2": year2, "period2": period2,
                        "year3": year3, "period3": period3,
                        "year4": year4, "period4": period4,
                    }
                else:
                    period_config = {
                        "year1": 2025, "period1": 6,
                        "year2": 2025, "period2": 3,
                        "year3": 2024, "period3": 12,
                        "year4": 2024, "period4": 9,
                    }
                print(f"ğŸ”„ {ticker} analiz ediliyor...")
                # Veri Ã§ekme
                api_data = await self.get_financial_data_api(session, ticker, period_config)
                market_data = await self.get_market_data(ticker, historical_date)
                # SektÃ¶r bilgisini belirle
                sector = "BILINMIYOR"
                for sec, stocks in self.sektoral_groups.items():
                    if ticker in stocks:
                        sector = sec
                        break
                if api_data and "value" in api_data and len(api_data["value"]) > 0:
                    # Veri iÅŸleme
                    df_data = []
                    for item in api_data["value"]:
                        row = {
                            "Kalem": item.get("itemDescTr", ""),
                            f"{period_config['year1']}-{period_config['period1']}": item.get("value1"),
                            f"{period_config['year2']}-{period_config['period2']}": item.get("value2"),
                            f"{period_config['year3']}-{period_config['period3']}": item.get("value3"),
                            f"{period_config['year4']}-{period_config['period4']}": item.get("value4")
                        }
                        df_data.append(row)
                    df = pd.DataFrame(df_data)
                    numeric_cols = [
                        f"{period_config['year1']}-{period_config['period1']}",
                        f"{period_config['year2']}-{period_config['period2']}",
                        f"{period_config['year3']}-{period_config['period3']}",
                        f"{period_config['year4']}-{period_config['period4']}"
                    ]
                    for col in numeric_cols:
                        df[col] = pd.to_numeric(df[col], errors='coerce') / 1000000
                    periods = numeric_cols
                    financial_data = {}
                    latest_period = None
                    for period in periods:
                        financial_data = self.extract_financial_values(df, period)
                        if financial_data.get('faaliyet_kari', 0) != 0:
                            latest_period = period
                            break
                    if not latest_period:
                        latest_period = periods[0]
                        logger.warning(f"{ticker} iÃ§in faaliyet kÃ¢rÄ± tÃ¼m dÃ¶nemlerde 0. VarsayÄ±lan dÃ¶nem: {latest_period}")
                    # Verileri birleÅŸtir
                    financial_data.update(market_data)
                    financial_data['sektor'] = sector
                    # Veri doÄŸrulama
                    warnings = self.validate_financial_data(financial_data)
                    # OranlarÄ± hesapla
                    ratios = self.calculate_ratios(financial_data)
                    financial_data.update(ratios)
                    # SkorlarÄ± hesapla
                    health_score = self.calculate_financial_health_score(financial_data)
                    valuation_score = self.calculate_valuation_score(financial_data)
                    future_score = self.calculate_future_performance_score(financial_data)
                    # Genel yatÄ±rÄ±m skoru (aÄŸÄ±rlÄ±klÄ± ortalama)
                    investment_score = (
                        health_score['score'] * 0.4 +
                        valuation_score['score'] * 0.35 +
                        future_score['score'] * 0.25
                    )
                    # SonuÃ§larÄ± oluÅŸtur
                    result = {
                        'HÄ°SSE KODU': ticker,
                        'SEKTÃ–R': sector,
                        'DÃ–NEM': latest_period,
                        'FÄ°YAT': round(market_data.get('current_price', 0), 2),
                        'EPS': round(market_data.get('eps', 0), 2),
                        'DEFTER DEÄERÄ°': round(market_data.get('book_value_per_share', 0), 2),
                        'NET SATIÅ(M)': round(financial_data.get('net_satis', 0), 1),
                        'NET KAR(M)': round(financial_data.get('net_kar', 0), 1),
                        'TOPLAM VARLIK(M)': round(financial_data.get('aktif_toplam', 0), 1),
                        'Ã–ZKAYNAK(M)': round(financial_data.get('ozsermaye', 0), 1),
                        'Ã–DENMÄ°Å SERMAYE(M)':round(financial_data.get('odenmis_sermaye',0),1),
                        'FAALÄ°YET KARI(M)': round(financial_data.get('faaliyet_kari', 0), 1),
                        'ROE (%)': round(ratios.get('roe', 0), 2),
                        'ROA (%)': round(ratios.get('roa', 0), 2),
                        'ROIC (%)': round(ratios.get('ROIC', 0), 2),
                        'NET.K.MARJ(%)': round(ratios.get('net_kar_marji', 0), 2),
                        'BRÃœT.K.MARJ(%)': round(ratios.get('brut_kar_marji', 0), 2),
                        'FAAL.K.MARJ(%)': round(ratios.get('faaliyet_kar_marji', 0), 2),
                        'EBITDA_MARJI': round(ratios.get('EBITDA_MARJI', 0), 2),
                        'BORÃ‡ ORAN(%)': round(ratios.get('borc_orani', 0), 2),
                        'LÄ°KÄ°T ORAN': round(ratios.get('likidite_orani', 0), 2),
                        'NAKÄ°T ORAN': round(ratios.get('nakit_orani', 0), 2),
                        'ASÄ°T.T.ORAN': round(ratios.get('asit_test', 0), 2),
                        'PD/DD': round(ratios.get('pd_dd', 0), 2),
                        'F/K': round(ratios.get('fk', 0), 2),
                        'EV/EBITDA': round(ratios.get('ev_ebitda', 0), 2),
                        'FÄ°YAT TRENDÄ° (%)': round(market_data.get('price_trend', 0), 2),
                        'SAÄLIK SKORU': health_score['score'],
                        'SAÄLIK KATEGORÄ°': health_score['category'],
                        'DEÄERLEME SKORU': valuation_score['score'],
                        'DEÄERLEME KATEGORÄ°': valuation_score['category'],
                        'GELECEK SKORU': future_score['score'],
                        'GELECEK KATEGORÄ°': future_score['category'],
                        'YATIRIM SKORU': round(investment_score, 1),
                        'YATIRIM KATEGORÄ°': self.get_score_category(investment_score),
                        'UYARILAR': "; ".join(warnings) if warnings else "Yok"
                    }

                    print(f"âœ… {ticker} tamamlandÄ± - YatÄ±rÄ±m Skoru: {result['YATIRIM SKORU']}/100")
                    return result
                else:
                    logger.error(f"{ticker} iÃ§in veri bulunamadÄ±")
                    return None
            except Exception as e:
                logger.error(f"{ticker} analiz hatasÄ±: {str(e)}")
                return None
    async def batch_analyze(self, tickers, batch_size=10, max_concurrent=3, historical_date=None):
        """Batch processing ile bÃ¼yÃ¼k veri setlerini analiz et"""
        all_results = []
        for i in range(0, len(tickers), batch_size):
            batch = tickers[i:i+batch_size]
            print(f"ğŸ”„ Batch {i//batch_size + 1}/{(len(tickers)-1)//batch_size + 1} iÅŸleniyor...")
            connector = aiohttp.TCPConnector(limit=20, limit_per_host=max_concurrent)
            timeout = aiohttp.ClientTimeout(total=60)
            async with aiohttp.ClientSession(connector=connector, timeout=timeout) as session:
                semaphore = asyncio.Semaphore(max_concurrent)
                tasks = [self.analyze_ticker(session, ticker, semaphore, historical_date) for ticker in batch]
                batch_results = await asyncio.gather(*tasks, return_exceptions=True)
                valid_results = [r for r in batch_results if r is not None and not isinstance(r, Exception)]
                all_results.extend(valid_results)
                # Ara sonuÃ§larÄ± kaydet
                self.save_intermediate_results(all_results)
                # Ã–nbelleÄŸi kaydet
                self.save_cache()
                print(f"âœ… Batch tamamlandÄ±: {len(valid_results)}/{len(batch)} hisse analiz edildi")

        # SektÃ¶r ortalamalarÄ±nÄ± hesaplama kÄ±smÄ±nÄ± kaldÄ±rdÄ±k
        # Her hisse iÃ§in hedef fiyat hesapla (artÄ±k sektÃ¶r ortalamalarÄ±nÄ± kullanmadan)
        for result in all_results:
            # SektÃ¶r ortalamalarÄ± yerine None deÄŸerleri gÃ¶nderiyoruz, fonksiyon iÃ§inde kullanÄ±lmayacak
            target_data = self.calculate_target_price(result, None, None)
            result.update(target_data)

        return all_results
    def save_intermediate_results(self, results):
        """Ara sonuÃ§larÄ± kaydet"""
        try:
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            filename = f"intermediate_results_{timestamp}.pkl"
            with open(filename, 'wb') as f:
                pickle.dump(results, f)
            logger.info(f"Ara sonuÃ§lar kaydedildi: {filename}")
        except Exception as e:
            logger.error(f"Ara sonuÃ§lar kaydedilemedi: {str(e)}")
    def display_results(self, results, top_n=20):
        """SonuÃ§larÄ± Colab'da gÃ¼zel formatta gÃ¶sterir"""
        if not results:
            print("âŒ GÃ¶sterilecek sonuÃ§ yok!")
            return
        df = pd.DataFrame(results)
        df = df.sort_values('YATIRIM SKORU', ascending=False).head(top_n)
        print("ğŸ“Š YATIRIM ANALÄ°Z SONUÃ‡LARI")
        print("=" * 80)
        print(f"ğŸ“ˆ En YÃ¼ksek YatÄ±rÄ±m Skoru: {df['YATIRIM SKORU'].max():.1f}/100 ({df.loc[df['YATIRIM SKORU'].idxmax(), 'HÄ°SSE KODU']})")
        print(f"ğŸ“‰ En DÃ¼ÅŸÃ¼k YatÄ±rÄ±m Skoru: {df['YATIRIM SKORU'].min():.1f}/100 ({df.loc[df['YATIRIM SKORU'].idxmin(), 'HÄ°SSE KODU']})")
        print(f"ğŸ“Š Ortalama YatÄ±rÄ±m Skoru: {df['YATIRIM SKORU'].mean():.1f}/100")
        print("\n")
        # HTML tablo oluÅŸtur
        html_table = df.to_html(classes='table table-striped', table_id='analysis_table', escape=False)
        # CSS stilleri
        css = """
        <style>
            #analysis_table {
                font-size: 12px;
                width: 100%;
                border-collapse: collapse;
            }
            #analysis_table th {
                background-color: #366092;
                color: white;
                padding: 8px;
                text-align: center;
            }
            #analysis_table td {
                padding: 6px;
                text-align: right;
            }
            #analysis_table tr:nth-child(even) {
                background-color: #f2f2f2;
            }
            .positive { color: green; font-weight: bold; }
            .negative { color: red; font-weight: bold; }
            .score-very-good { background-color: #d4edda; }
            .score-good { background-color: #d1ecf1; }
            .score-medium { background-color: #fff3cd; }
            .score-weak { background-color: #f8d7da; }
            .score-very-weak { background-color: #f5c6cb; }
        </style>
        """
        display(HTML(css + html_table))
        return df
    def plot_analysis(self, results):
        """Analiz sonuÃ§larÄ±nÄ± gÃ¶rselleÅŸtir"""
        if not results:
            print("âŒ GÃ¶rselleÅŸtirilecek veri yok!")
            return
        df = pd.DataFrame(results)
        # 1. YatÄ±rÄ±m Skoru DaÄŸÄ±lÄ±mÄ±
        plt.figure(figsize=(12, 8))
        plt.subplot(2, 2, 1)
        sns.histplot(df['YATIRIM SKORU'], bins=20, kde=True, color='skyblue')
        plt.title('YatÄ±rÄ±m Skoru DaÄŸÄ±lÄ±mÄ±')
        plt.xlabel('YatÄ±rÄ±m Skoru')
        # 2. SektÃ¶rel KarÅŸÄ±laÅŸtÄ±rma
        plt.subplot(2, 2, 2)
        sector_avg = df.groupby('SEKTÃ–R')['YATIRIM SKORU'].mean().sort_values(ascending=False).head(10)
        sns.barplot(x=sector_avg.values, y=sector_avg.index, palette='viridis')
        plt.title('SektÃ¶rel YatÄ±rÄ±m SkorlarÄ± (Ä°lk 10)')
        plt.xlabel('Ortalama YatÄ±rÄ±m Skoru')
        # 3. F/K vs ROE
        plt.subplot(2, 2, 3)
        scatter = plt.scatter(
            x=df['F/K'],
            y=df['ROE (%)'],
            s=df['YATIRIM SKORU']*2,
            c=df['YATIRIM SKORU'],
            cmap='RdYlGn',
            alpha=0.7
        )
        plt.xlabel('F/K OranÄ±')
        plt.ylabel('ROE (%)')
        plt.title('F/K vs ROE (Boyut = YatÄ±rÄ±m Skoru)')
        plt.colorbar(scatter, label='YatÄ±rÄ±m Skoru')
        # 4. Hedef Fiyat vs Mevcut Fiyat
        plt.subplot(2, 2, 4)
        plt.scatter(
            x=df['FÄ°YAT'],
            y=df['HEDEF FÄ°YAT'],
            c=df['YÃœKSELÄ°Å POTANSÄ°YELÄ° (%)'],
            cmap='RdYlGn',
            s=100,
            alpha=0.7
        )
        plt.plot([df['FÄ°YAT'].min(), df['FÄ°YAT'].max()],
                [df['FÄ°YAT'].min(), df['FÄ°YAT'].max()],
                'r--', alpha=0.5)
        plt.xlabel('Mevcut Fiyat')
        plt.ylabel('Hedef Fiyat')
        plt.title('Hedef Fiyat vs Mevcut Fiyat')
        plt.colorbar(label='YÃ¼kseliÅŸ Potansiyeli (%)')
        plt.tight_layout()
        plt.show()
        # En iyi 10 hisse iÃ§in detaylÄ± grafik
        top_10 = df.nlargest(10, 'YATIRIM SKORU')
        plt.figure(figsize=(14, 8))
        metrics = ['SAÄLIK SKORU', 'DEÄERLEME SKORU', 'GELECEK SKORU']
        x = np.arange(len(top_10))
        width = 0.25
        for i, metric in enumerate(metrics):
            plt.bar(x + i*width, top_10[metric], width, label=metric)
        plt.xlabel('Hisse Senetleri')
        plt.ylabel('Skor')
        plt.title('En Ä°yi 10 Hisse iÃ§in DetaylÄ± Skorlar')
        plt.xticks(x + width, top_10['HÄ°SSE KODU'], rotation=45)
        plt.legend()
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
    def export_to_excel(self, results, filename="yatirim_analizi.xlsx", backtest_results=None):
        """SonuÃ§larÄ± Excel'e aktar"""
        if not results:
            print("âŒ Excel'e aktarÄ±lacak sonuÃ§ yok!")
            return
        df = pd.DataFrame(results)
        df = df.sort_values('YATIRIM SKORU', ascending=False)
        # SÃ¼tun sÄ±rasÄ±nÄ± dÃ¼zenle
        columns = [
            'HÄ°SSE KODU', 'SEKTÃ–R', 'FÄ°YAT', 'HEDEF FÄ°YAT', 'YÃœKSELÄ°Å POTANSÄ°YELÄ° (%)', 'DÃ–NEM',
            'YATIRIM SKORU', 'YATIRIM KATEGORÄ°', 'SAÄLIK SKORU', 'SAÄLIK KATEGORÄ°',
            'DEÄERLEME SKORU', 'DEÄERLEME KATEGORÄ°', 'GELECEK SKORU', 'GELECEK KATEGORÄ°',
            'EPS', 'DEFTER DEÄERÄ°', 'NET SATIÅ(M)', 'NET KAR(M)', 'TOPLAM VARLIK(M)',
            'Ã–ZKAYNAK(M)','Ã–DENMÄ°Å SERMAYE(M)','FAALÄ°YET KARI(M)', 'ROE (%)', 'ROA (%)', 'ROIC (%)',
            'NET.K.MARJ(%)', 'BRÃœT.K.MARJ(%)', 'FAAL.K.MARJ(%)', 'EBITDA_MARJI',
            'BORÃ‡ ORAN(%)', 'LÄ°KÄ°T ORAN', 'NAKÄ°T ORAN', 'ASÄ°T.T.ORAN', 'PD/DD', 'F/K',
            'EV/EBITDA', 'FÄ°YAT TRENDÄ° (%)', 'HEDEF FÄ°YAT AÃ‡IKLAMA', 'UYARILAR'
        ]
        df = df[columns]
        # Excel'e yaz
        with pd.ExcelWriter(filename, engine='openpyxl') as writer:
            # Ana analiz sayfasÄ±
            df.to_excel(writer, sheet_name='YATIRIM ANALÄ°ZÄ°', index=False)
            # SektÃ¶rel analiz
            sector_analysis = []
            for sector in df['SEKTÃ–R'].unique():
                sector_data = df[df['SEKTÃ–R'] == sector]
                if not sector_data.empty:
                    sector_summary = {
                        'SEKTÃ–R': sector,
                        'HÄ°SSE SAYISI': len(sector_data),
                        'ORT. YATIRIM SKORU': sector_data['YATIRIM SKORU'].mean(),
                        'ORT. SAÄLIK SKORU': sector_data['SAÄLIK SKORU'].mean(),
                        'ORT. DEÄERLEME SKORU': sector_data['DEÄERLEME SKORU'].mean(),
                        'ORT. GELECEK SKORU': sector_data['GELECEK SKORU'].mean(),
                        'ORT. YÃœKSELÄ°Å POTANSÄ°YELÄ°': sector_data['YÃœKSELÄ°Å POTANSÄ°YELÄ° (%)'].mean(),
                        'ORT. ROE': sector_data['ROE (%)'].mean(),
                        'ORT. F/K': sector_data['F/K'].mean(),
                        'ORT. PD/DD': sector_data['PD/DD'].mean()
                    }
                    sector_analysis.append(sector_summary)
            pd.DataFrame(sector_analysis).to_excel(writer, sheet_name='SEKTÃ–REL ANALÄ°Z', index=False)
            # En iyi hisseler
            top_investment = df.nlargest(20, 'YATIRIM SKORU')[
                ['HÄ°SSE KODU', 'SEKTÃ–R', 'YATIRIM SKORU', 'FÄ°YAT', 'HEDEF FÄ°YAT', 'YÃœKSELÄ°Å POTANSÄ°YELÄ° (%)', 'F/K', 'PD/DD', 'ROE (%)']
            ]
            top_investment.to_excel(writer, sheet_name='EN Ä°YÄ° 20 YATIRIM', index=False)
            # Ucuz hisseler
            cheap_stocks = df[df['DEÄERLEME SKORU'] >= 70].nlargest(20, 'YATIRIM SKORU')[
                ['HÄ°SSE KODU', 'SEKTÃ–R','Ã–DENMÄ°Å SERMAYE(M)', 'DEÄERLEME SKORU', 'FÄ°YAT', 'HEDEF FÄ°YAT', 'YÃœKSELÄ°Å POTANSÄ°YELÄ° (%)', 'F/K', 'PD/DD', 'ROE (%)']
            ]
            cheap_stocks.to_excel(writer, sheet_name='EN UCUZ 20 HÄ°SSE', index=False)
            # SaÄŸlÄ±klÄ± hisseler
            healthy_stocks = df[df['SAÄLIK SKORU'] >= 70].nlargest(20, 'YATIRIM SKORU')[
                ['HÄ°SSE KODU', 'SEKTÃ–R', 'SAÄLIK SKORU', 'FÄ°YAT', 'HEDEF FÄ°YAT', 'YÃœKSELÄ°Å POTANSÄ°YELÄ° (%)', 'ROE (%)', 'BORÃ‡ ORAN(%)', 'LÄ°KÄ°T ORAN']
            ]
            healthy_stocks.to_excel(writer, sheet_name='EN SAÄLIKLI 20 HÄ°SSE', index=False)
            # YÃ¼ksek potansiyelli hisseler
            high_potential = df[df['YÃœKSELÄ°Å POTANSÄ°YELÄ° (%)'] > 20].nlargest(20, 'YATIRIM SKORU')[
                ['HÄ°SSE KODU', 'SEKTÃ–R', 'YÃœKSELÄ°Å POTANSÄ°YELÄ° (%)', 'FÄ°YAT', 'HEDEF FÄ°YAT', 'YATIRIM SKORU', 'F/K', 'PD/DD', 'ROE (%)']
            ]
            high_potential.to_excel(writer, sheet_name='YÃœKSEK POTANSÄ°YELLÄ° 20', index=False)




        # Formatlama
        self.format_excel_advanced(filename)
        print(f"âœ… Excel dosyasÄ± oluÅŸturuldu: {filename}")
        # Colab iÃ§in indirme
        try:
            from google.colab import files
            files.download(filename)
        except:
            print("âš ï¸ Dosya otomatik olarak indirilemedi. LÃ¼tfen manuel olarak indirin.")
    def format_excel_advanced(self, filename):
        """Excel dosyasÄ±nÄ± profesyonel formatta dÃ¼zenler"""
        try:
            wb = openpyxl.load_workbook(filename)
            # TÃ¼m sayfalar iÃ§in formatlama
            for sheet_name in wb.sheetnames:
                ws = wb[sheet_name]
                # BaÅŸlÄ±k formatÄ±
                header_font = Font(name='Arial', size=10, bold=True, color="FFFFFF")
                header_fill = PatternFill(start_color="2F5597", end_color="2F5597", fill_type="solid")
                header_alignment = Alignment(horizontal="center", vertical="center")
                # Veri formatÄ±
                data_font = Font(name='Arial', size=9)
                data_alignment = Alignment(horizontal="right", vertical="center")
                # BaÅŸlÄ±klarÄ± formatla
                for cell in ws[1]:
                    cell.font = header_font
                    cell.fill = header_fill
                    cell.alignment = header_alignment
                # Verileri formatla
                for row in ws.iter_rows(min_row=2):
                    for i, cell in enumerate(row):
                        cell.font = data_font
                        # SayÄ±sal sÃ¼tunlarÄ± formatla
                        if isinstance(cell.value, (int, float)):
                            if 'SKORU' in str(ws.cell(row=1, column=i+1).value):
                                cell.number_format = '#,##0'
                            elif '%' in str(ws.cell(row=1, column=i+1).value):
                                cell.number_format = '#,##0.00'
                            elif i >= 15:  # Finansal veriler
                                cell.number_format = '#,##0.0'
                            else:
                                cell.number_format = '#,##0.00'
                            cell.alignment = data_alignment
                        else:
                            cell.alignment = Alignment(horizontal="center", vertical="center")
                # SÃ¼tun geniÅŸlikleri
                if sheet_name == 'YATIRIM ANALÄ°ZÄ°':
                    column_widths = {
                        'A': 12,  # HÄ°SSE KODU
                        'B': 25,  # SEKTÃ–R
                        'C': 15,  # ANALÄ°Z TARÄ°HÄ° (varsa)
                        'D': 10,  # FÄ°YAT
                        'E': 12,  # HEDEF FÄ°YAT
                        'F': 18,  # YÃœKSELÄ°Å POTANSÄ°YELÄ°
                        'G': 10,  # DÃ–NEM
                        'H': 15,  # YATIRIM SKORU
                        'I': 18,  # YATIRIM KATEGORÄ°
                        'J': 15,  # SAÄLIK SKORU
                        'K': 18,  # SAÄLIK KATEGORÄ°
                        'L': 18,  # DEÄERLEME SKORU
                        'M': 20,  # DEÄERLEME KATEGORÄ°
                        'N': 18,  # GELECEK SKORU
                        'O': 20,  # GELECEK KATEGORÄ°
                        'P': 12,  # EPS
                        'Q': 15,  # DEFTER DEÄERÄ°
                        'R': 20,  # NET SATIÅ
                        'S': 15,  # NET KAR
                        'T': 20,  # TOPLAM VARLIK
                        'U': 15,  # Ã–ZKAYNAK
                        'V': 20,  # FAALÄ°YET KARI
                        'W': 10,  # ROE
                        'X': 10,  # ROA
                        'Y': 12,  # ROIC
                        'Z': 15,  # NET KAR MARJI
                        'AA': 15, # BRÃœT KAR MARJI
                        'AB': 15, # FAALÄ°YET KAR MARJI
                        'AC': 15, # EBITDA MARJI
                        'AD': 15, # BORÃ‡ ORANI
                        'AE': 15, # LÄ°KÄ°T ORAN
                        'AF': 15, # NAKÄ°T ORAN
                        'AG': 15, # ASÄ°T TEST
                        'AH': 12, # PD/DD
                        'AI': 10, # F/K
                        'AJ': 12, # EV/EBITDA
                        'AK': 15, # FÄ°YAT TRENDÄ°
                        'AL': 50, # HEDEF FÄ°YAT AÃ‡IKLAMA
                        'AM': 50  # UYARILAR
                    }
                    # EÄŸer ANALÄ°Z TARÄ°HÄ° sÃ¼tunu varsa diÄŸer sÃ¼tunlarÄ± kaydÄ±r
                    if 'ANALÄ°Z TARÄ°HÄ°' in [cell.value for cell in ws[1]]:
                        for col in list(column_widths.keys())[3:]:
                            column_widths[col] = column_widths[chr(ord(col)-1)]
                    for col, width in column_widths.items():
                        if col in ws.column_dimensions:
                            ws.column_dimensions[col].width = width
                # KoÅŸullu formatlama - YatÄ±rÄ±m Skoru
                if sheet_name == 'YATIRIM ANALÄ°ZÄ°':
                    investment_col = None
                    for i, cell in enumerate(ws[1], 1):
                        if cell.value == 'YATIRIM SKORU':
                            investment_col = openpyxl.utils.get_column_letter(i)
                            break
                    if investment_col:
                        # Renk skalasÄ±
                        color_scale = ColorScaleRule(
                            start_type='min', start_color='FF0000',
                            mid_type='percentile', mid_value=50, mid_color='FFFF00',
                            end_type='max', end_color='00FF00'
                        )
                        ws.conditional_formatting.add(f'{investment_col}2:{investment_col}{ws.max_row}', color_scale)
            wb.save(filename)
            print("âœ… Excel formatlarÄ± uygulandÄ±")
        except Exception as e:
            print(f"âš ï¸ Excel formatlama hatasÄ±: {str(e)}")
# Analyzer ve BackTest instance'larÄ± oluÅŸtur
analyzer = AdvancedFinancialAnalyzer()
print("âœ… GeliÅŸmiÅŸ Finansal Analiz hazÄ±r!")

# =============================================================================
# 3. HÄ°SSE LÄ°STESÄ° TANIMALAMA
# =============================================================================
bist500_tickers = ["A1CAP","A1YEN","ACSEL","ADEL","ADESE","ADGYOâ€‹","AEFESâ€‹","AFYONâ€‹","AGESA","AGHOL","AGROT","AGYO","AHGAZ","AKBNK","AKCNS","AKENR","AKFGY","AKFIS","AKFYE",
"AKGRT","AKMGY","AKSA","AKSEN","AKSGY","AKSUE","AKYHO","ALARK","ALBRK","ALCAR","ALCTL","ALFAS","ALGYO","ALKA","ALKIM","ALKLC","ALMAD","ALTINY","ALVES","ANELE","ANGEN","ANHYT",
"ANSGR","ARASE","ARCLK","ARDYZ","ARENA","ARMGD","ARSAN","ARTMS","ARZUM","ASELS","ASGYO","ASTOR","ASUZU","ATAGY","ATAKP","ATATP","ATEKS","ATLAS","ATSYH","AVGYO","AVHOL","AVOD",
"AVPGY","AVTUR","AYCES","AYDEM","AYEN","AYES","AYGAZ","AZTEK","BAGFS","BAHKM","BAKAB","BALAT","BALSU","BANVT","BARMA","BASCMâ€‹","BASGZ","BAYRK","BEGYO","BERA","BEYAZ","BFREN",
"BIENY","BIGCH","BIMAS","BINBN","BINHO","BIOEN","BIZIM","BJKAS","BLCYT","BMSCH","BMSTL","BNTAS","BOBET","BORLS","BORSK","BOSSA","BRISA","BRKO","BRKSN","BRKVY","BRLSM",
"BRMEN","BRSAN","BRYAT","BSOKE","BTCIM","BUCIM","BURCE","BURVA","BVSAN","BYDNR","CANTE","CASA","CATES","CCOLA","CELHA","CEMAS","CEMTS","CEOEM","CEMZY","CGCAM",
"CIMSAâ€‹","CLEBI","CMBTN","CMENT","CONSE","COSMO","CRDFA","CRFSA","CUSAN","CVKMD","CWENE","DAGHL","DAGI","DAPGM","DARDL","DENGE","DERHL","DERIM","DESA","DCTTR","DSTFK",
"DESPC","DEVA","DGATE","DGGYO","DGNMO","DIRIT","DITAS","DMRGD","DMSAS","DNISI","DOAS","DOBUR","DOCO","DOFER","DOGUB","DOHOL","DOKTA","DURDO","DYOBY","DURKN",
"DZGYO","EBEBK","ECILC","ECZYT","EDATA","EDIP","EFORC","EGEGY","EGEEN","EGEPO","EGGUB","EGPRO","EGSER","EKGYO","EKIZ","EKOS","EKSUN","ELITE","EMKEL","EMNIS","ENERY","ENDEA",
"ENJSA","ENKAI","ENSRI","ENTRA","EPLAS","ERBOS","ERCB","EREGL","ERSU","ESCAR","ESCOM","ESEN","ETILR","ETYAT","EUHOL","EUKYO","EUPWR","EUREN","EUYO","EYGYO",
"FADE","FENER","FLAP","FMIZP","FONET","FORMT","FORTE","FRIGO","FROTO","FZLGY","GARAN","GARFA","GEDIK","GEDZA","GENIL","GENTS","GEREL","GESAN","GIPTA",
"GLBMD","GLCVY","GLRMK","GLRYH","GLYHO","GMTAS","GOKNR","GOLTS","GOODY","GOZDE","GRNYO","GRSEL","GRTHO","GRTRK","GSDDE","GSDHOâ€‹","GSRAY","GUBRF","GUNDG","GWIND","GZNMI","HALKB",
"HATEK","HATSN","HDFGS","HEDEF","HEKTS","HKTM","HLGYO","HOROZ","HRKET","HTTBT","HUBVC","HUNER","HURGZ","ICBCTâ€‹","ICUGS","ICUGS","IDGYO","IEYHO","IHAAS","IHEVAâ€‹",
"IHGZT","IHLAS","IHLGM","IHYAY","IMASM","INDES","INFO","INGRM","INTEMâ€‹","INVEO","INVES","IPEKE","ISATR","ISBIR","â€‹ISCTR","ISDMR","ISFIN","ISGSY","ISGYO",
"ISKPL","ISKUR","ISMEN","ISSEN","ISYAT","IZENR","IZFAS","IZINV","IZMDC","JANTS","KAPLM","KAREL","KARSN","KARTN","KARYE","KATMR","KAYSE","KBORU","KCAER",
"KCHOL","KENT","KERVN","KERVTâ€‹","KFEIN","KGYO","KIMMR","KLGYO","KLKIM","KLMSN","KLNMA","KLRHO","KLSER","KLSYN","KLYPV","KMPUR","KNFRT","KOCMT","KONKA","KONTR","KONYA",
"KOPOL","KORDS","KOTON","KOZAA","KOZAL","KRDMA","KRDMB","KRDMD","KRGYO","KRONT","KRPLS","KRSTL","KRTEK","KRVGD","KSTUR","KTLEV","KTSKR","KUTPO","KUVVA","KUYAS",
"KZBGY","KZGYO","LIDER","LIDFA","LILAK","LINK","LKMNH","LOGO","LRSHO","LUKSK","LYDHO","LYDYE","MAALT","MACKO","MAGEN","MAKIM","MAKTK","MANAS","MARBL","MARKA","MARTI","MAVI",
"MEDTR","MEGAP","MEGMT","MEKAG","MEPET","MERCN","MERIT","MERKOâ€‹","METRO","METUR","MGROS","MHRGY","MIATK","MIPAZ","MMCAS","MNDRS","MNDTR","MOBTL","MPARK","MOBTL","MOGAN","MOPAS",
"MRGYO","MRSHL","MSGYO","MTRKS","MTRYO","MZHLD","NATEN","NETAS","NIBAS","NTGAZ","NTHOL","NUGYO","NUHCM","OBAMS","OBASE","ODAS","ODINE","OFSYM","ONCSM","ONRYT", "ORCAY","ORGE",
"ORMA","OSMEN","OSTIM","OTKAR","OTTOâ€‹","OYAKC","OYAYO","OYLUM","OYYAT","OZATD","OZGYO","OZKGY","OZRDN","OZSUB","OZYSR","PAGYO","PAMEL","PAPIL","PARSN","PASEU","PCILT","PATEK","PEHOL",
"PEKGY","PENGD","PENTA","PETKM","PETUN","PGSUS","PINSU","PKART","PKENT","PLTUR","PNLSN","PNSUTâ€‹","POLHO","POLTK","PRDGS","PRKAB","PRKME","PRZMA",
"PSDTC","PSGYO","QNBFB","QNBFL","QUAGR","RALYH","RAYSG","REEDR","RGYAS","RNPOL","RODRG","ROYAL","RTALB","RUBNS","RUZYE","RYGYO","RYSAS","SAFKR","SAHOL","SAMAT","SANEL",
"SANFM","SANKO","SARKY","SASA","SAYAS","SDTTR","SEGMN","SEGYOâ€‹","SEKFK","SEKUR","SELEC","SELGD","SELVA","SERNT","SEYKMâ€‹","SILVR","SISE","SKBNK","SKTAS","SKYLP","SKYMD",
"SMARTâ€‹","SMRTG","SNGYO","SNICA","SNKRN","SNPAM","SODSN","SOKE","SOKM","SONME","SRVGY","SUMAS","SUNTK","SURGY","SUWEN","TABGD","TARKM","TATEN","TATGD",
"TAVHL","TBORG","TCELL","TCKRC","TDGYO","TEKTU","TERA","TETMT","TEZOL","TGSAS","THYAO","TKFEN","TKNSA","TLMAN","TMPOL","TMSN","TNZTPâ€‹","TOASO","TRCAS","TRGYO",
"TRILC","TSGYO","TSKB","TSPOR","TTKOM","TTRAK","TUCLK","TUKAS","TUPRS","TUREX","TURGG","TURSG","UFUK","ULAS","ULKER","ULUFA","ULUSE","ULUUN","UMPAS",
"UNLUâ€‹","USAK","UZERB","VAKBN","VAKFN","VAKKO","VANGDâ€‹","VBTYZ","VERTU","VERUS","VESBE","VESTL","VKFYO","VKGYO","VKING","VSNMD","YAPRK","YATAS","YAYLA","YBTAS",
"YEOTK","YESIL","YGGYO","YGYOâ€‹","YIGIT","YKBNK","YKSLN","YONGA","YUNSA","YYAPI","YYLGD","ZEDUR","ZOREN","ZRGYO"
]
bist30_tickers = [
    "THYAO", "AKBNK", "GARAN", "ISCTR", "KCHOL", "TUPRS", "ARCLK", "BIMAS",
    "SAHOL", "KOZAL", "SISE", "TAVHL", "TCELL", "TKFEN", "TOASO", "TRKCM",
    "VESTL", "YKBNK", "PGSUS", "DOHOL", "EKGYO", "ENKAI", "FROTO", "HALKB",
    "KONTR", "KRDMD", "MGROS", "PETKM", "VAKBN", "AEFES"
]
print("ğŸ“‹ Hisse listeleri hazÄ±r!")
print(f"ğŸ”¬ BIST30 listesi: {len(bist30_tickers)} hisse")
print(f"ğŸ”¬ BIST500 listesi: {len(bist500_tickers)} hisse")
# =============================================================================
# 4. HIZLI TEST BÃ–LÃœMÃœ
# =============================================================================
print("\nğŸš€ HIZLI TEST BAÅLIYOR...")
print("=" * 50)
quick_test = ["THYAO", "MACKO", "SASA"]
async def quick_analysis():
    results = await analyzer.batch_analyze(bist500_tickers, batch_size=3, max_concurrent=3)
    return results
test_results = await quick_analysis()
if test_results:
    print("\nâœ… HIZLI TEST BAÅARILI!")
    analyzer.display_results(test_results)
    analyzer.plot_analysis(test_results)
    analyzer.export_to_excel(test_results, "hisse_analizi.xlsx")
else:
    print("âŒ Test baÅŸarÄ±sÄ±z - API baÄŸlantÄ±sÄ±nÄ± kontrol edin")

# =============================================================================
# 6. KULLANIM TALÄ°MATLARI
# =============================================================================
print("""
ğŸ¯ KULLANIM TALÄ°MATLARI:
========================
1ï¸âƒ£ KÃœÃ‡ÃœK TEST Ä°Ã‡Ä°N:
   results = await analyzer.batch_analyze(bist30_tickers[:10], batch_size=5, max_concurrent=2)
   analyzer.display_results(results)
   analyzer.plot_analysis(results)
   analyzer.export_to_excel(results, "kucuk_test.xlsx")
2ï¸âƒ£ BÃœYÃœK ANALÄ°Z Ä°Ã‡Ä°N:
   results = await analyzer.batch_analyze(bist30_tickers, batch_size=10, max_concurrent=3)
   analyzer.display_results(results, top_n=30)
   analyzer.plot_analysis(results)
   analyzer.export_to_excel(results, "buyuk_analiz.xlsx")
3ï¸âƒ£ BÄ°ST500 TAM ANALÄ°Z Ä°Ã‡Ä°N:
   results = await analyzer.batch_analyze(bist500_tickers, batch_size=20, max_concurrent=5)
   analyzer.display_results(results, top_n=50)
   analyzer.plot_analysis(results)
   analyzer.export_to_excel(results, "bist500_tam_analiz.xlsx")
4ï¸âƒ£ Ã–ZEL HÄ°SSE LÄ°STESÄ° Ä°Ã‡Ä°N:
   my_stocks = ["THYAO", "SASA", "KCHOL", "TUPRS", "ARCLK"]
   results = await analyzer.batch_analyze(my_stocks, batch_size=5, max_concurrent=2)
   analyzer.display_results(results)
   analyzer.plot_analysis(results)
   analyzer.export_to_excel(results, "ozel_listem.xlsx")
5ï¸âƒ£ TEK HÄ°SSE DETAYLI ANALÄ°Z Ä°Ã‡Ä°N:
   # Ã–nce analiz et
   results = await analyzer.batch_analyze(["THYAO"], batch_size=1, max_concurrent=1)
   # SonuÃ§larÄ± gÃ¶ster
   if results:
       df = pd.DataFrame(results)
       display(df.T)  # Transpoz gÃ¶sterim
       # DetaylÄ± skorlarÄ± yazdÄ±r
       print("ğŸ“Š SAÄLIK SKORU DETAYLARI:")
       for key, value in results[0]['SAÄLIK SKORU']['details'].items():
           print(f"  {key}: {value}")
       print("\\nğŸ“Š DEÄERLEME SKORU DETAYLARI:")
       for key, value in results[0]['DEÄERLEME SKORU']['details'].items():
           print(f"  {key}: {value}")
       print("\\nğŸ“Š GELECEK SKORU DETAYLARI:")
       for key, value in results[0]['GELECEK SKORU']['details'].items():
           print(f"  {key}: {value}")
       print("\\nğŸ¯ HEDEF FÄ°YAT BÄ°LGÄ°LERÄ°:")
       print(f"  Mevcut Fiyat: {results[0]['FÄ°YAT']} TL")
       print(f"  Hedef Fiyat: {results[0]['HEDEF FÄ°YAT']} TL")
       print(f"  YÃ¼kseliÅŸ Potansiyeli: %{results[0]['YÃœKSELÄ°Å POTANSÄ°YELÄ° (%)']}")
       print(f"  AÃ§Ä±klama: {results[0]['HEDEF FÄ°YAT AÃ‡IKLAMA']}")

- Ä°lk Ã§alÄ±ÅŸtÄ±rmada kÃ¼Ã§Ã¼k listeyle baÅŸlayÄ±n
- Ã–nbellekleme sayesinde tekrar eden analizler hÄ±zlÄ±dÄ±r
- Rate limiting nedeniyle bÃ¼yÃ¼k listeler zaman alabilir
- Backtest iÃ§in geÃ§miÅŸ veriler kullanÄ±lÄ±r, bu nedenle sonuÃ§lar geÃ§miÅŸ performansa dayanÄ±r
- Mobil uygulama iÃ§in bu kod bir API backend olarak kullanÄ±labilir
ğŸ¯ YATIRIM SKORU ANLAMI:
- 80-100: Ã‡OK Ä°YÄ° (GÃ¼Ã§lÃ¼ alÄ±m sinyali)
- 60-79: Ä°YÄ° (AlÄ±m dÃ¼ÅŸÃ¼nÃ¼lebilir)
- 40-59: ORTA (NÃ¶tr, izlenmeli)
- 20-39: ZAYIF (SatÄ±ÅŸ dÃ¼ÅŸÃ¼nÃ¼lebilir)
- 0-19: Ã‡OK ZAYIF (KaÃ§Ä±nÄ±lmalÄ±)
ğŸ“ˆ DEÄERLEME ORANLARI DOÄRU MANTIÄI:
- F/K: DÃ¼ÅŸÃ¼k = Ucuz, YÃ¼ksek = PahalÄ±
- PD/DD: DÃ¼ÅŸÃ¼k = Ucuz, YÃ¼ksek = PahalÄ±
- EV/EBITDA: DÃ¼ÅŸÃ¼k = Ucuz, YÃ¼ksek = PahalÄ± """)
```
